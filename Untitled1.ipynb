```python
import random
from random import *
```

```python
class Fighter:
    
    name = "Fighter"
    health = 35
    phy_dmg = 0
    crit_bon = False
    is_alive = True
    
    def __init__(self,name):
        
        self.name = name
    
    def attacks(self):
        criti = randint(1,11)
        print(f"{self.name} attacks!")
        if criti > 8:
            print("The powers of the Mushroom lend a hand. Critical damage incoming!")
            self.crit_bon = True
            
    def takes_damage(self):
        h1 = self.health
        dmg = randint(1,11) 
        
        self.health -= dmg
        
        if self.name == fr_player.name:
            if sc_player.crit_bon == True:
                self.health -= 5
                sc_player.crit_bon = False
        else:
            if fr_player.crit_bon == True:
                self.health -= 5
                fr_player.crit_bon = False    
                 
        print(f"{self.name} takes {h1-self.health} damage!")
        if h1 - self.health > dmg:
            print(f"Total Damage: {dmg} + Crit Bonus")
        else:
            print(f"Total Damage: {dmg}")
        
        print(f"{self.name}'s health is down to: {self.health}")
```

```python
choice = input("Mario, would you like to go first?").lower()

if choice == "yes":
    fr_player = Fighter("Mario")
    sc_player = Fighter("Luigi")
else:
    fr_player = Fighter("Luigi")
    sc_player = Fighter("Mario")

while fr_player.is_alive == True and sc_player.is_alive == True:
    fr_player.attacks()
    sc_player.takes_damage()
    
    if sc_player.health < 1:
        print(f"{sc_player.name} is out for the count!")
        sc_player.is_alive = False
    if fr_player.health - sc_player.health > 8 and sc_player.is_alive == True:
        print(f"The fight's not over, but it's not looking good for {sc_player.name}!!!")
    
    if fr_player.is_alive == True and sc_player.is_alive == True:
        sc_player.attacks()
        fr_player.takes_damage()
        
        if fr_player.health < 1:
            print(f"{fr_player.name} is out for the count!")
            fr_player.is_alive = False
        if sc_player.health - fr_player.health > 8 and fr_player.is_alive == True:
            print(f"The fight's not over, but it's not looking good for {fr_player.name}!!!")
            
print("GAME OVER!!!")
fr_player = Fighter("Mario")
sc_player = Fighter("Luigi")
```
